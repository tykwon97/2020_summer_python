````python
class 캐릭터():
    def __init__(self):
        """캐릭터의 속성으로, HP, SP, 무기를 가진다."""
        
    def 일반공격(self):
        """캐릭터는 일반공격으로 몬스터의 HP를 감소시킨다."""
        
    def 특수공격(self):
        """캐릭터는 특수공격으로 몬스터의 HP를 감소시키고, 자신의 SP가 감소한다."""
````

````python
class 전사(캐릭터):
    def __init__(self):
        """전사의 속성으로, HP=100, SP=50, 무기를 가진다."""
        self.HP = 1000
        self.SP = 50
        self.무기 = None
        self.num = 5 #치명타를 때릴 기본 확률
        self.exp = 0 # 경험치
        self.best_score=0 #최고 스코어를 항상 보여주면 갱신할 경우 바꾸기
        self.levelup=0
    
    def 일반공격(self, 공격대상):
        
        """전사는 일반공격으로 몬스터의 HP를 40만큼 감소시킨다."""
        if(randint(1, self.num) == 1) : #치명타를 맞은 경우
                공격대상.HP -= 50
                print(f"**********치명타를 주었습니다**********")
                print(f"일반 공격, 몬스터의 남은 HP: {공격대상.HP}")
                self.exp += 15
                print(f"현재 경험치 : {self.exp}")

        else :
                공격대상.HP -= 30
                print(f"일반 공격, 몬스터의 남은 HP: {공격대상.HP}")
                self.exp += 10
                print(f"현재 경험치 : {self.exp}")
        
    def 특수공격(self, 공격대상):
        
        
        """전사는 특수공격으로 몬스터의 HP를 100만큼 감소시키고, 자신의 SP가 50만큼 감소한다."""
        if self.SP >= 50:
            if(randint(1, self.num) == 1) : #치명타를 때린 경우 : 추가 데미지로 처리
                공격대상.HP -= 10
                print(f"**********치명타를 주었습니다**********")
            self.SP -= 50
            공격대상.HP -= 100
            print(f"특수 공격, 몬스터의 남은 HP: {공격대상.HP}")
            self.exp += 20
            print(f"현재 경험치 : {self.exp}")
        else:
            print(f"SP가 부족합니다. 현재 SP: {self.SP}")
        
    def 무기선택(self):
        """캐릭터는 무기를 선택할 수 있습니다."""
        print("무기를 선택합니다.\n")
        무기목록 = {"녹슨칼": "녹슨칼", "강철검" : "강철검", "야구배트": "야구배트"}
        print("""녹슨 칼   : HP +50 SP -50 치명타 +3
강철 검   : HP -50 SP +50 치명타 +1
야구 배트 : HP -30 SP +30 치명타 +2
                """)
        무기 = input("""
        녹슨칼, 강철검, 야구배트 중 무기를 선택하세요.
        """)
        while(무기 not in 무기목록) :
            print("존재하지 않는 무기입니다. 다시 입력해주세요.")
            무기 = input("""
        녹슨칼, 강철검, 야구배트 중 무기를 선택하세요.
        """)
        self.무기 = 무기목록[무기]
        if(self.무기 == "녹슨칼"):
            self.SP -= 50
            self.HP += 50
            self.num -= 3
        elif(self.무기 == "강철검"):
            self.SP += 50
            self.HP -= 50
            self.num -= 1
        else:
            self.SP += 30
            self.HP -= 30
            self.num -= 2
        return print(f"{self.무기}를(을) 선택하셨습니다.\n")
    
    def 전직(self):
        self.SP += 100
        self.HP += 150
        print("************************다음 단계로 전직했습니다***********************.")
        print("SP가 100 HP가 150 증가했습니다.")
        return print("*****************************************************************\n")
    
````

몬스터

````python
class 몬스터():
    def __init__(self):
        """몬스터의 속성으로 HP를 가진다."""
        
    def 일반공격(self):
        """몬스터는 캐릭터에게 데미지를 입힌다."""
````

````python
class 버섯괴물(몬스터):
    def __init__(self):
        """버섯괴물의 속성으로 HP를 100가진다."""
        self.HP = 100
        
    def 일반공격(self, 공격대상):
        """몬스터는 캐릭터에게 데미지를 입힌다."""
        
        # 공격대상이 살아있고, 몬스터도 살아있을 때 공격을 할 수 있습니다.
        while 공격대상.HP > 0 and self.HP > 0:
            sleep(3)
            if(randint(1, 5) == 1) : #치명타를 맞은 경우
                공격대상.HP -= 50
                print(f"**********치명타를 맞았습니다**********")
                print(f"버섯괴물이 공격을 합니다. 캐릭터 HP: {공격대상.HP}")
            else :
                공격대상.HP -= 30
                print(f"버섯괴물이 공격을 합니다. 캐릭터 HP: {공격대상.HP}")
                
````

````python
class 멧돼지(몬스터):
    def __init__(self):
        """멧돼지의 속성으로 HP를 150가진다."""
        self.HP = 150
        
    def 일반공격(self, 공격대상):
        """몬스터는 캐릭터에게 데미지를 입힌다."""
        
        # 공격대상이 살아있고, 몬스터도 살아있을 때 공격을 할 수 있습니다.
        while 공격대상.HP > 0 and self.HP > 0:
            sleep(3)
            if(randint(1, 3) == 1) : #치명타를 맞은 경우
                공격대상.HP -= 70
                print(f"**********치명타를 맞았습니다**********")
                print(f"멧돼지가 공격을 합니다. 캐릭터 HP: {공격대상.HP}")

            else :
                공격대상.HP -= 50
                print(f"멧돼지가 공격을 합니다. 캐릭터 HP: {공격대상.HP}")
            
````

장면

````python
class 장면():
    def 입장(self):
        """장면의 속성은 캐릭터나 몬스터의 객체이다."""
        
    def 다음장면(self):
        """장면은 다음 장면으로 넘어가는 기능을 가지고 있다."""
````

````python
class 시작(장면):
    def 입장(self):
        """캐릭터 선택을 하게 하고 객체를 생성한 뒤 이 객체를 마을로 넘긴다."""
        print("게임을 시작합니다. ")
        self.캐릭터 = self.캐릭터_선택()
        return self.다음장면()
        
    def 캐릭터_선택(self):
        """사용자의 입력을 받아 캐릭터 객체를 생성한다."""
        
        캐릭터_목록 = {"전사" : 전사(),
                       "마법사" : 마법사()}
        
        캐릭터 = input("전사와 마법사 중 원하는 캐릭터를 선택해주세요.: ")
        while(캐릭터 not in 캐릭터_목록) :
            print("존재하지 않는 캐릭터입니다. 다시 입력해주세요.")
            캐릭터 = input("전사와 마법사 중 원하는 캐릭터를 선택해주세요.: ")
        캐릭터 = 캐릭터_목록[캐릭터]
        return 캐릭터
        
    def 다음장면(self):
        """캐릭터 선택이 끝나면 다음 장면인 마을로 캐릭터 객체를 전달한다."""
        print("캐릭터 직업 선택이 완료되었습니다. 마을로 이동합니다.")
        다음장면 = 마을(self.캐릭터) # 이걸 받아줄 마을 클래스의 생성자가 필요하다.
        return 다음장면.입장() # 마을 객체
````

````python
class 마을(장면):
    def __init__(self, 캐릭터):
        self.캐릭터 = 캐릭터
    
    def 입장(self):
        """마을에서는 캐릭터의 무기를 선택하게 한다."""
        print("마을에 입장했습니다.")
        self.캐릭터.무기선택()
        self.다음장면()
        
    def 다음장면(self):
        """무기 선택이 끝나면 다음 장면인 동굴로 캐릭터 객체를 전달한다."""
        
        print("마을에서 무기 선택을 완료했습니다. 버섯사냥터로 이동합니다.")
        다음장면 = 버섯사냥터(self.캐릭터)
        return 다음장면.입장()
    
    def 엔딩(self):
        """캐릭터가 모든 미션을 마치고 마을로 돌아와 게임을 종료한다."""
        return print("마을로 복귀, 게임 끝!")
````

### @staticmethod가 무엇인가요?


```
위 표현은 정적메서드를 표현하기 위함입니다.

우리는 지금까지 인스턴스 속성, 인스턴스 메서드만 사용했습니다. 즉, 객체를 만들어야만 해당 속성과 메서드를 사용할 수 있었습니다.

정적 메서드는, 객체 생성을 하지 않아도 바로 사용할 수 있는 메서드입니다.

'정적'이라는 단어가 의미하듯, 정적 메서드에는 다른 메서드나 속성에 영향을 받지 않고 순수하게 해당 기능만 수행할 필요한 때에 사용합니다.

경우에 따라서 정적메서드가 필요한 경우도 있기 때문에 참고하면 좋을 것 같습니다.

자세한 내용이 궁금하다면 아래 링크를 참고해보시면 될 것 같습니다.
```

````python
class 전사(캐릭터):
    def __init__(self):
        """전사의 속성으로, HP=100, SP=50, 무기를 가진다."""
        self.HP = 100
        self.SP = 50
        self.무기 = None
    
    def 일반공격(self, 공격대상):
        """전사는 일반공격으로 몬스터의 HP를 40만큼 감소시킨다."""
        공격대상.HP -= 40
        print(f"일반 공격, 몬스터의 남은 HP: {공격대상.HP}")
        
    def 특수공격(self, 공격대상):
        """전사는 특수공격으로 몬스터의 HP를 100만큼 감소시키고, 자신의 SP가 50만큼 감소한다."""
        if self.SP >= 50:
            self.SP -= 50
            공격대상.HP -= 100
            print(f"특수 공격, 몬스터의 남은 HP: {공격대상.HP}")
        else:
            print(f"SP가 부족합니다. 현재 SP: {self.SP}")
        
    def 무기선택(self):
        """캐릭터는 무기를 선택할 수 있습니다."""
        print("무기를 선택합니다.")
        무기목록 = {"녹슨칼": "녹슨칼", "강철검" : "강철검", "야구배트": "야구배트"}
        
        무기 = input("""
        녹슨칼, 강철검, 야구배트 중 무기를 선택하세요.
        """)
        self.무기 = 무기목록[무기]
        return print(f"{self.무기}를(을) 선택하셨습니다.")
````

### 버섯괴물과 멧돼지 클래스가 thread 클래스와 몬스터 클래스를 상속하게 만듭니다.

````python
class 버섯괴물(threading.Thread, 몬스터):
    def __init__(self, 공격대상):
        """버섯괴물의 속성으로 HP를 100가진다."""
        threading.Thread.__init__(self)
        super(버섯괴물, self).__init__()
        self.HP = 100
        self.공격대상 = 공격대상
        
    def run(self):
        """몬스터는 캐릭터에게 데미지를 입힌다."""
        while self.공격대상.HP > 0 and self.HP > 0:
            self.공격대상.HP -= 5
            print(f"버섯괴물이 공격을 합니다. 캐릭터 HP: {self.공격대상.HP}")
            sleep(3)
````

````python
class 멧돼지(threading.Thread, 몬스터):
    def __init__(self, 공격대상):
        """멧돼지의 속성으로 HP를 150가진다."""
        threading.Thread.__init__(self)
        super(멧돼지, self).__init__()
        self.HP = 150
        self.공격대상 = 공격대상
        
    def run(self):
        """몬스터는 캐릭터에게 데미지를 입힌다."""
        while self.공격대상.HP > 0 and self.HP > 0:
            self.공격대상.HP -= 7
            print(f"멧돼지가 공격을 합니다. 캐릭터 HP: {self.공격대상.HP}")
            sleep(3)
````

````python
class 장면():
    def 입장(self):
        """장면의 속성은 캐릭터나 몬스터의 객체이다."""
        
    def 다음장면(self):
        """장면은 다음 장면으로 넘어가는 기능을 가지고 있다."""
````

````python
class 시작(장면):
    def 입장(self):
        """캐릭터 선택을 하게 하고 객체를 생성한 뒤 이 객체를 마을로 넘긴다."""
        print("게임을 시작합니다. ")
        self.캐릭터 = self.캐릭터_선택()
        return self.다음장면()
        
    def 캐릭터_선택(self):
        """사용자의 입력을 받아 캐릭터 객체를 생성한다."""
        
        캐릭터_목록 = {"전사" : 전사(),
                       "마법사" : 마법사()}
        
        캐릭터 = input("전사와 마법사 중 원하는 캐릭터를 선택해주세요.: ")
        캐릭터 = 캐릭터_목록[캐릭터]
        return 캐릭터
        
    def 다음장면(self):
        """캐릭터 선택이 끝나면 다음 장면인 마을로 캐릭터 객체를 전달한다."""
        print("캐릭터 직업 선택이 완료되었습니다. 마을로 이동합니다.")
        다음장면 = 마을(self.캐릭터) # 이걸 받아줄 마을 클래스의 생성자가 필요하다.
        return 다음장면.입장() # 마을 객체
````

````python
class 마을(장면):
    def __init__(self, 캐릭터):
        self.캐릭터 = 캐릭터
    
    def 입장(self):
        """마을에서는 캐릭터의 무기를 선택하게 한다."""
        print("마을에 입장했습니다.")
        self.캐릭터.무기선택()
        self.다음장면()
        
    def 다음장면(self):
        """무기 선택이 끝나면 다음 장면인 동굴로 캐릭터 객체를 전달한다."""
        print("마을에서 무기 선택을 완료했습니다. 동굴로 이동합니다.")
        다음장면 = 버섯사냥터(self.캐릭터)
        return 다음장면.입장()
    
    def 엔딩(self):
        """캐릭터가 모든 미션을 마치고 마을로 돌아와 게임을 종료한다."""
        return print("마을로 복귀, 게임 끝!")
````

````python
class 버섯사냥터(장면):
    def __init__(self, 캐릭터):
        self.캐릭터 = 캐릭터
        self.몬스터 = 버섯괴물(self.캐릭터)
    
    def 입장(self):
        """버섯사냥터에서는 캐릭터와 버섯괴물이 대결한다. 캐릭터가 이기면 다음 장면으로 이동하고, 버섯괴물이 이기면 게임을 재시작하거나 종료한다."""
        print("""
        버섯사냥터에서 버섯괴물을 발견했습니다. 마을의 평화를 위해 버섯괴물을 해치워 주세요.
        """)
        self.몬스터.start()
        
        while self.몬스터.HP > 0:
            액션 = input("""행동을 선택하세요.
                            1. 일반 공격
                            2. 특수 공격""")
            if (액션 == "일반 공격") or (액션 == "1"):
                self.캐릭터.일반공격(self.몬스터)
            elif (액션 == "특수 공격") or (액션 == "2"):
                self.캐릭터.특수공격(self.몬스터)
        self.다음장면()
        
    def 다음장면(self):
        """캐릭터가 버섯괴물을 죽이면 다음 장면인 돼지사냥터으로 캐릭터 객체를 전달한다."""    
        다음장면 = 돼지사냥터(self.캐릭터)
        return 다음장면.입장()
````

````python
class 돼지사냥터(장면):
    def __init__(self, 캐릭터):
        self.캐릭터 = 캐릭터
        self.몬스터 = 멧돼지(self.캐릭터)
    
    def 입장(self):
        """돼지사냥터에서는 캐릭터와 멧돼지가 대결한다. 캐릭터가 이기면 다음 장면으로 이동하고, 버섯괴물이 이기면 게임을 재시작하거나 종료한다.""""""동굴에서는 캐릭터와 버섯괴물이 대결한다. 캐릭터가 이기면 다음 장면으로 이동하고, 멧돼지가 이기면 게임을 재시작하거나 종료한다."""
        print("""
        돼지사냥터에서 멧돼지를 발견했습니다. 마을의 평화를 위해 멧돼지를 해치워 주세요.
        """)
        self.몬스터.start()        
        
        while self.몬스터.HP > 0:
            액션 = input("""행동을 선택하세요.
                            1. 일반 공격
                            2. 특수 공격: """)
            if (액션 == "일반 공격") or (액션 == "1"):
                self.캐릭터.일반공격(self.몬스터)
            elif (액션 == "특수 공격") or (액션 == "2"):
                self.캐릭터.특수공격(self.몬스터)
        self.다음장면()
        
    def 다음장면(self):
        """캐릭터가 멧돼지를 죽이면 마을로 돌아와 엔딩을 한다."""
        다음장면 = 마을(self.캐릭터)
        return 다음장면.엔딩()
````
