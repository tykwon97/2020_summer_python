배달음식 메뉴의 총합과 주문목록을 구하려고 합니다.

다음의 조건에 맞는 코드를 완성하세요.

`클래스명: Order`<br>
`인스턴스 속성: order_num, menu_list`<br>
`인스턴스 메서드: total_price, total_menu`<br>

`클래스명: Menu`<br>
`인스턴스 속성: name, price`

```python     

# 메뉴 객체 생성        
치킨 = Menu(name = "치킨", price = 18000)
짜장면 = Menu(name = "짜장면", price = 10000)
피자 = Menu(name = "피자", price = 21000)

# 메뉴를 합쳐서 전달
menu_list = [치킨, 짜장면, 피자]

new_order = Order(order_num = 555, menu_list=menu_list)

# 주문 결과 출력

# 결과

주문합계 49000원입니다.
주문목록은 ['치킨', '짜장면', '피자']입니다.
```




````python
# write your code here!!   

class order :
    def __init__(self,order_num,menu_list):
        self.order_num=order_num
        self.menu_list=menu_list
        self.total=0 # 주문 합계
        self.list=[] # 주문 목록
        
    def total_price(self):
        for i in (self.menu_list) :         
            self.total += int(i.price) # 각각의 가격을 하나씩 더해준다
        return self.total
    
    def total_menu(self):
        for i in (self.menu_list) :         
            self.list.append(i.name) #주문 목록에 하나씩 추가하기
        return (self.list)
    
class menu :
    def __init__ (self,name,price) :
        self.name=name
        self.price=price
    
치킨 = menu(name = "치킨", price = 18000)
짜장면 = menu(name = "짜장면", price = 10000)
피자 = menu(name = "피자", price = 21000)    

menu_list = [치킨, 짜장면, 피자]
new_order = order(order_num = 555, menu_list=menu_list)

print(f"주문합계 {new_order.total_price()}원입니다.")
print(f"주문목록은 {new_order.total_menu()}입니다.")
````

# Q6

김철수씨는 최근에 어느 광산에 취직했습니다.

그 광산엔 금, 은, 다이아몬드가 있다고 합니다.

김철수씨의 목표는 금, 은, 다이아몬드를 각각 1개씩 채굴하는 것이고

김철수씨를 광부 클래스의 객체로 생성하고 위의 문제를 해결하세요.


```python

# 실행예시

# 목표=[금,은,다이아몬드] 순서로 대입합니다. 즉, 아래 목표는 금, 은, 다이아를 각각 1개씩으로 하는 것을 목표로 한 것입니다.

김철수 = 광부(이름='김철수', 목표=[1,1,1])

김철수.채굴()
    무슨 광물을 채굴할까? 금
    뚝딱뚝뚝....(3초)
    금을(를) 채굴했습니다.
    
김철수.광물확인()
    김철수씨는 현재 금 1개, 은 0개, 다이아몬드 0개를 가지고 있습니다.
    목표까지 금 0개, 은 1개, 다이아몬드 1개가 필요합니다.
    
김철수.채굴()
    무슨 광물을 채굴할까? 은
    뚝딱뚝뚝....(3초)
    은을(를) 채굴했습니다.
    
김철수.채굴()
    무슨 광물을 채굴할까? 다이아몬드
    뚝딱뚝뚝....(3초)
    다이아몬드을(를) 채굴했습니다.
    
김철수.광물확인()
    김철수씨는 현재 금 1개, 은 1개, 다이아몬드 1개를 가지고 있습니다.
    축하합니다. 목표를 달성했습니다!
    
```

````python
# write your code here!!

# sleep(n) 을 하면 n초간 코드 진행을 지연할 수 있습니다.

#금,은,다이아몬드를 한꺼번에 배열에 넣어서 활용해도 된다.

from time import sleep

import random

class 광부 :
    def __init__ (self,이름,목표):
        self.이름=이름
        self.목표=목표
        self.gold=0 # 금              
        self.silver=0 #은
        self.diamond=0 #다이아몬드
        self.num=0 #랜덤값 저장 변수
        self.count=self.목표[0]+self.목표[1]+self.목표[2] #총 보물의 개수
    def 채굴(self) :
        if(self.count>0) :
            while(True) :
                self.num=random.randrange(0,3) #랜덤 값 생성
                if (self.목표[self.num] !=0) : #해당 보물을 모두 채취하지 않은 경우
                    self.목표[self.num] -= 1
                    self.count -= 1 # 총 보물의 개수만큼 반복
                    break
            if(self.num==0): # 금인 경우
                self.gold += 1 #채취한 금 갯수 증가
                print("무슨 광물을 채굴할까? 금")
                print("뚝딱뚝뚝....")
                sleep(3) #3초동안 멈춘다.
                print("금을(를) 채굴했습니다.\n")
            elif(self.num==1): # 은인 경우
                self.silver += 1 #채취한 은 갯수 증가
                print("무슨 광물을 채굴할까? 은")
                print("뚝딱뚝뚝....")
                sleep(3)
                print("은을(를) 채굴했습니다.\n")
            else: # 다이아몬드인 경우
                self.diamond += 1 #채취한 다이아몬드 갯수 증가
                print("무슨 광물을 채굴할까? 다이아몬드")
                print("뚝딱뚝뚝....")
                sleep(3)
                print("다이아몬드을(를) 채굴했습니다.\n")
        
    def 광물확인(self) :
        if(self.count>0) : #아직 채취할 보물이 남은 경우
            print(f"김철수씨는 현재 금 {self.gold}개, 은 {self.silver}개, 다이아몬드 {self.diamond}개를 가지고 있습니다.")
            print(f"목표까지 금 {self.목표[0]}개, 은 {self.목표[1]}개, 다이아몬드 {self.목표[2]}개가 필요합니다.\n")
        else : # 모두 채취한 경우
            print(f"김철수씨는 현재 금 {self.gold}개, 은 {self.silver}개, 다이아몬드 {self.diamond}개를 가지고 있습니다.")
            print("축하합니다. 목표를 달성했습니다!\n")
김철수 = 광부(이름='김철수', 목표=[1,1,1])
김철수.채굴()
김철수.광물확인()
김철수.채굴()
김철수.채굴()
김철수.광물확인()
````

단, 잔액이 부족하다면 출금이 불가능합니다.<br>
save, withdraw, check_balance는 각각 입금, 출금, 잔액 확인을 하는 메서드임


````python
# write your code here!!

#int형으로 입력받을때 변경시켜주면 편하다.

class atm :
    def __init__(self,money) :
        self.money=money #기본 금액
        self.count=0 #입금 출금
        self.total=self.money #총합
        while(True) :#4번이 입력되기전까지 무한 반복
            self.num=input("1. 입금\n2. 출금\n3. 잔액확인\n4. 거래종료\n선택(1~4) :")
            if(int(self.num)==1) : #1번이 입력된 경우 : 
                self.save() #save 메소드 실행
            elif(int(self.num)==2) : #2번이 입력된 경우 : 
                self.withdraw() #withdraw 메소드 실행
            elif(int(self.num)==3) : #3번이 입력된 경우 : 
                self.check_balance() #check_balance메소드 실행
            elif(int(self.num)==4) : #4번이 입력된 경우 : 
                self.exit() #exit 메소드 실행
                break #반복 종료
            else :
                print("벗어난 범위입니다.")
                
    def save(self) :
        self.count=input("입금액 : ") #입금액 입력 받기
        self.total += int(self.count) #입력받은 금액은 string형 int형 변형해야됨
        print("잔액 : ",self.total)
    def withdraw(self) :
        self.count=input("출금액 : ") #출금액 입력 받기
        self.total -= int(self.count) #입력받은 금액은 string형 int형 변형해야됨
        if(self.total<0) : #출금액이 잔액보다 큰 경우
            print("잔액이 부족합니다.")
            self.total += int(self.count)
        print("잔액 : ",self.total)
    def check_balance(self) :
        print("잔액 : ",self.total)
    def exit(self) :
        print("거래를 종료합니다.")
    
김철수 = atm(money = 10000)


    
````

````python
# write your code here!!
class bank :
    def __init__(self,name,balance) :
        self.name=name
        self.balance=balance
    def transfer(self,target,money) :
        if((target.balance-money)>0) :#송금 금액이 잔액보다 작은 경우
            self.balance -= money # 잔액에서 뺀다
            target.balance += money #송금금액에 추가
            print(f"{self.name}가 {target.name}에게 {money}원을 송금했습니다. {self.name}의 잔액은 {self.balance}입니다.")
        else :
            print("잔액이 부족하여 송금이 불가능합니다.")
    def check_balance (self):
        print(f"{self.name}의 잔액은 {self.balance}입니다.")
        
철수 = bank(name = '철수', balance = 10000)
짱구 = bank(name = '짱구', balance = 20000)
철수.transfer(target = 짱구, money = 2000)
짱구.check_balance()
철수.transfer(target = 짱구, money = 50000)
````

어느 게임에서 초보자는 마법사로 전직할 수 있다고 합니다.

마법사는 초보자의 모든 속성을 중복하여 가지고 있다고 합니다.

`상속을 이용해서 중복된 코드를 제거해주세요.`

````python
# write your code here!!

class 초보자:
    def __init__(self, name, hp):
        self.name = name
        self.hp = hp
        
    def print_info(self):
        return print(f"이름: {self.name} HP: {self.hp}")
    
    
class 마법사(초보자):#초보자 클래스 상속하기
    def __init__(self,name,hp, sp):
        super().__init__(name,hp) #초보자 클래스 가져오기
        self.sp = sp
        
    def print_info(self):
        return print(f"이름: {self.name} HP: {self.hp}, SP: {self.sp}")
    
    
wizard_obj = 마법사("닥터스트레인지", 100, 100)
wizard_obj.print_info()
````

````python
# write your code here!!
class human :
    def __init__(self,name,phone) :
        self.name=name
        self.phone=phone
    def print_info(self) :
        print("이름 : ",self.name)
        print("번호 : ",self.phone)
class student(human) :
    def __init__(self,name,phone,college,student_id):
        super().__init__(name,phone) #human 클래스 상속하기
        self.college=college
        self.student_id=student_id
    def print_info(self) :
        print("이름 : ",self.name)
        print("번호 : ",self.phone)
        print("학교 : ",self.college)
        print("학번 : ",self.student_id)
        
철수 = student(name = "철수", phone= "010-1234-1234", college = "Harvard", student_id = 1234)
철수.print_info()
````
